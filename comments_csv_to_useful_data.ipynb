```python
import json
import ast
import pandas as pd

# Read the CSV file
comments_df = pd.read_csv('@tiktok_video_7106594312292453675_comments.csv')

def process_comment(comment_i):
    # Convert None strings to actual None values
    def _convert_none_strings(d):
        if isinstance(d, dict):
            return {k: _convert_none_strings(v) for k, v in d.items()}
        elif isinstance(d, list):
            return [_convert_none_strings(x) for x in d]
        elif d == 'None':
            return None
        return d
    
    row = comments_df.iloc[comment_i]
    user_dict = ast.literal_eval(comments_df['user'][0])
    user_dict = _convert_none_strings(user_dict)

    commenter_id = user_dict['uid']
    comment_text = row['text']

    likes_on_this_comment = row['digg_count']
    is_liked_by_author = row['is_author_digged']
    comment_timestamp = row['create_time']

    return {
        'commenter_id': commenter_id,
        'text': comment_text,
        'likes_on_this_comment': likes_on_this_comment,
        #'is_liked_by_author': is_liked_by_author,
        'timestamp': comment_timestamp
    }

# Process all comments into list of dictionaries
comments_list = []
for i in range(len(comments_df)):
    try:
        comment_dict = process_comment(i)
        comments_list.append(comment_dict)
    except Exception as e:
        print(f"Error processing comment {i}: {e}")
        continue

print(f"Successfully processed {len(comments_list)} comments")
```
